# 5 types of flowers classification using resnet50

importing the necessasary libraries

```python
from fastai.vision.all import *
from pathlib import Path
```

getting the path for the images.

```python
path=Path('/kaggle/input/5-flower-types-classification-dataset')
```

```python
# defining the datablock
block=DataBlock(
blocks=(ImageBlock,CategoryBlock),
get_items=get_image_files,
get_y=parent_label,
item_tfms=Resize(224))
#dataloaders
loader=block.dataloaders(path)
#batch of data
loader.show_batch()
```

# Classification task

```python
#creating the classifier/learner
cls=vision_learner(loader,resnet50,metrics=error_rate)
```

# Finetuning the model

```python
cls.fine_tune(10)
```

```python
cls.show_results()
```

```python
metrics=cls.validate()
print(metrics)
acc=1-metrics[0]
print(f"the accuracy is {acc}")
```

# the accuracy is 88.38%

```python
interpret=ClassificationInterpretation.from_learner(cls)
interpret.plot_confusion_matrix()
```

saving the model

```python
cls.export('/kaggle/working/class.pkl')
```
